```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.utils import to_categorical
from matplotlib import pyplot as plt
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
y_train = to_categorical(y_train,10)
y_test = to_categorical(y_test,10)
```

```python
for i in range(9):
	# define subplot
	plt.subplot(330 + 1 + i)
	# plot raw pixel data
	plt.imshow(x_train[i])
```

```python
print(x_train.shape)
print(x_train[0])
```

```python
for i in range(9):
	# define subplot
	plt.subplot(330 + 1 + i)
	# plot raw pixel data
	plt.imshow(x_test[i])
```

```python
print(x_test.shape)
print(x_test[0])
```

```python
from re import VERBOSE
Samkhya = Sequential()
Samkhya.add(Flatten())
Samkhya.add(Dense(10, activation='softmax'))
Samkhya.compile(loss='categorical_crossentropy')
Samkhya.fit(x_train,y_train,epochs=1,verbose=1)
print(Samkhya.evaluate(x_test,y_test,verbose=1))
plt.imshow(x_test[0])
print(np.argmax(Samkhya.predict(x_test[0:1]),axis=1))
```
